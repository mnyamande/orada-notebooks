```python
# Create synthetic dataset
animal_classes = ['dog', 'cat', 'horse', 'elephant']
X = np.random.rand(100, 100)  # Simulated image features
y = np.random.choice(animal_classes, 100)
```

```python
# Split training and testing set
X_train, X_test = X[:80], X[80:]
y_train, y_test = y[:80], y[80:]
```

```python
# Split and train model
model = keras.Sequential([
    layers.Dense(64, activation='relu', input_shape=(100,)),
    layers.Dense(32, activation='relu'),
    layers.Dense(len(animal_classes), activation='softmax')
])

model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

model.fit(X_train,
          np.array([animal_classes.index(cls) for cls in y_train]),
          epochs=10)
```

```python
# Save model
model.save('animal_classifier.keras')
with open('animal_classes.pkl', 'wb') as f:
    pickle.dump(animal_classes, f)
```
