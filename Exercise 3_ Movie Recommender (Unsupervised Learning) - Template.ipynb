```python
# Collaborative Filtering Recommender
import numpy as np
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity
import pickle
```

```python
# Simulated Movie User Preferences
movies = {
    'action': ['Avengers', 'John Wick', 'Mad Max'],
    'comedy': ['Hangover', 'Deadpool', 'Superbad'],
    'scifi': ['Inception', 'Matrix', 'Interstellar']
}

user_preferences = {
    'action_lover': [1, 1, 0],  # Likes action movies
    'comedy_fan': [0, 1, 1],    # Likes comedy movies
    'scifi_nerd': [0, 0, 1]     # Likes sci-fi movies
}

def recommend_movie(user_pref):
    # Compute similarity and recommend
    similarities = cosine_similarity([user_pref],
                                     list(user_preferences.values()))[0]
    most_similar_user = list(user_preferences.keys())[np.argmax(similarities)]
    genre = most_similar_user.split('_')[0]
    recommendation = np.random.choice(movies[genre])
    return recommendation
```

```python
# Save recommender function
with open('movie_recommender.pkl', 'wb') as f:
    pickle.dump(recommend_movie, f)
```
